# Logistic Regression Assignment

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
trainData = pd.read_csv("/content/drive/MyDrive/ML: Codes and Assignments/Week 2 - Assignments (Linear and Logistic Regression)/Logistic Regression/Training Data/Logistic_X_Train.csv")
trainLabels = pd.read_csv("/content/drive/MyDrive/ML: Codes and Assignments/Week 2 - Assignments (Linear and Logistic Regression)/Logistic Regression/Training Data/Logistic_Y_Train.csv")
```

```
Xtrain = trainData.values
Xtrain.shape
```

```
Ytrain = trainLabels.values.flatten()
Ytrain.shape
```

```
def hypothesis(x,w,b):
    hx = np.dot(x,w)+b
    return sigmoid(hx)

def sigmoid(hx):
    return 1.0/(1.0 + np.exp(-1.0*hx))
```

```
def gradAscent(x, y, learning_rate = 0.5, iter = 1000):
    
    m = x.shape[0]
    w = np.zeros(x.shape[1])
    b = 0.0
    likelihood = []
    for i in range(iter):
      hx = hypothesis(x, w, b)
      loglike = 1/m * np.sum(y*np.log(hx) + (1 - y)*np.log(1 - hx))
      likelihood.append(loglike)
      grad_w = 1/m * np.dot(x.T, (y - hx))
      grad_b = 1/m * np.sum(y - hx)
      w += learning_rate * grad_w
      b += learning_rate * grad_b
    return likelihood, w, b
```

```
likelihood, w, b = gradAscent(Xtrain, Ytrain)
```

```
plt.plot(likelihood)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter3D(f2, f3, f1, cmap = 'hsv')
# plt.show()
xt = f2
yt = f3
zt = -((w[1]*xt + w[2]*yt + b)/w[0])

ax.plot3D(xt,yt,zt,color = 'k',alpha = 0.2)
plt.show()
```
